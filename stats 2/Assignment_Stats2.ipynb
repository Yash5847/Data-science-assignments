<h4> Question4:- </h4>

```python
#Two values, a and b, symmetric about the mean, such that the 
#probability of the random variable taking a value between them is 0.99.
#From the above details,we have to exclude .005% area from each
#left and right tails. Hence, we want to find the .005th and the 
#.995th percentiles Z score values

mean = 100
std = 20

# Z value for .005 percentiles 
z_005_ = np.round(st.norm.ppf(0.005),4)
z_005_

# Z value for .99 percentiles 
z_99_ = np.round(st.norm.ppf(0.995),4)
z_99_

#z = (x_bar - mew) / std
#x_bar = (z*std) + mew
a = np.round((z_005_*std) + mean,1)
b = np.round((z_99_*std) + mean,1)
print("The two values of a and b, symmetric about the mean are such that the probability of the random variable taking a value between them is 0.99:",a,b)
```

<h4> Question5(A):- </h4>

```python
# Considering 1 USD = 45 rupees
# Combine Mean Profit of both division for Company= mean1 + mean2

mean1 = 5
mean2 = 7
Mean = (mean1+mean2) 

print('The Mean Profit for both division:',Mean, 'Million$')
print()
print('The Mean Profit for both division:',(Mean*45)/10, 'Crore Rupees')
print()
# Combine standard Deviation = (Std1^2 + Std2^2)^1/2

std1 = 3**2
std2 = 4**2
Std = np.sqrt(std1 + std2)

print('The Standard Deviation for both division:', Std, 'Million$')
print()
print('The Standard Deviation for both division:', (Std*45)/10, 'Crore Rupees')
```

```python
x1, x2 = np.round(st.norm.interval(0.95, Mean, Std),2)
print('Rupee Ranges from',x1,'to',x2,'Million$ in Annual profit of the Company 95% of the time')
print()
print('Rupee Ranges from',np.divide(np.multiply(x1,45),10),'to',np.divide(np.multiply(x2,45),10),
      'Crore Rupees in Annual profit of the Company 95% of the time')
```

<h4> Question5(B):- </h4>

```python
Z_05 = st.norm.ppf(0.05)
percentage_val = (Z_05 * Std) + Mean
print('The 5th percentile of Profit for the company is',np.round(percentage_val,2),'Million$')
print()
print('The 5th percentile of Profit for the company is',np.round((percentage_val*45)/10,),'Crore Rupees')
```
